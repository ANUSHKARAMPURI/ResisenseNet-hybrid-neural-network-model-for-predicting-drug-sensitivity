```python
import numpy as np

def sentences2vec_with_ids(sentences, model, unseen=None):
    """
    Generate vectors for each sentence (list of IDs) in a list of sentences. The vector is
    simply the sum of vectors for individual IDs (molecular identifiers).

    Parameters:
    - sentences: List of lists, where each list contains molecular IDs (integers).
    - model: Pretrained word2vec model (from gensim).
    - unseen: Optional value for words (IDs) not found in the model (default is None).

    Returns:
    - np.array: Array of vectors corresponding to each sentence (list of IDs).
    """
    # Get the vocabulary keys from the model (IDs in this case)
    keys = set(model.wv.key_to_index.keys())  # Gensim 4.x update
    
    vec = []
    if unseen:
        unseen_vec = model.wv[unseen]

    # Loop over each sentence (which is a list of IDs)
    for sentence in sentences:
        sentence_vec = []
        for word in sentence:  # Directly iterate over the list of IDs
            if word in keys:
                sentence_vec.append(model.wv[word])
            else:
                if unseen:
                    sentence_vec.append(unseen_vec)
        
        # Sum the vectors of the sentence and append it to the final list
        vec.append(np.sum(sentence_vec, axis=0))
    
    return np.array(vec)
```

PROTEIN EMBEDDING

METHOD 1 USING TRANSFORMERS - BertTokenizer, BertModel

Embedding with Pre-trained Models (e.g., ProtBERT, ESM-1b):

These are models trained specifically for protein and amino acid sequences. They generate embeddings for sequences using neural networks similar to how word embeddings work in NLP. They produce dense, continuous vector representations for sequences.

    ProtBERT: A BERT-based model pre-trained on protein sequences that provides embeddings for amino acid sequences.
    ESM-1b: A model that provides embeddings for protein sequences and works well for downstream tasks like classification or regression.

You can use HuggingFace's transformers library to use these models:

```python
from transformers import BertTokenizer, BertModel

# Load the pre-trained ProtBERT model
tokenizer = BertTokenizer.from_pretrained("Rostlab/prot_bert_bfd")
model = BertModel.from_pretrained("Rostlab/prot_bert_bfd")

# Example sequence
sequence = "ACDEFGHIKLMNPQRSTVWY"

# Tokenize the sequence
inputs = tokenizer(sequence, return_tensors="pt")

# Get the embeddings (from the last layer)
outputs = model(**inputs)
embeddings = outputs.last_hidden_state

# The embeddings for each amino acid in the sequence
print(embeddings)
```

```python
embeddings
```

```python
# Convert the tensor to a NumPy array for easier manipulation
embedding_array = embeddings.detach().numpy()

# The shape of the array will be (1, sequence_length, embedding_dim)
# We can ignore the batch dimension (1) and focus on the sequence length and embedding dimension
embedding_array = embedding_array[0]

# Now, embedding_array has shape (sequence_length, embedding_dim), 
# and each row represents the embedding of one amino acid in the sequence
print(embedding_array)
```

```python
embedding_array_df = pd.DataFrame(embedding_array)
```

```python
import numpy as np

# Assume 'embedding_array' is your 3x1024 NumPy array
average_embedding = np.mean(embedding_array, axis=0)

# Now average_embedding will be a 1D array of length 1024, representing the full sequence
print(average_embedding)
```
